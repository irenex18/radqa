```python
import pandas as pd
import numpy as np
import re
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metrics = pd.read_csv('radqa_data1.csv')
metrics.head()
```

```python
metrics.columns
```

```python
base_losses = ['Training Loss Init Params', 'Validation Loss Init Params', 'Training Loss Adjusted Params', 'Validation Loss Adjusted Params']
general_losses = ['Training Loss', 'Validation Loss']
losses = [base_losses, general_losses]

for loss in losses:
    plt.figure(figsize=(10, 6))
    for column in loss: 
        plt.plot(metrics['Step'], metrics[column], label=column)

    # Set labels and title
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.title('Evaluation Metrics for each Model')

    # Display legend
    plt.legend()

    # Show the plot
    plt.show()
```

```python
metrics.columns
```

```python
base_rouges = ['Rouge1 Init Params', 'Rouge2 Init Params', 'Rougel Init Params','Rouge1 Adjusted Params', 'Rouge2 Adjusted Params','Rougel Adjusted Params']
general_rouges = ['Rouge1','Rouge2', 'Rougel']
rouges = [base_rouges, general_rouges]

for rouge in rouges:
    plt.figure(figsize=(10, 6))
    for column in rouge: 
        plt.plot(metrics['Step'], metrics[column], label=column)

    # Set labels and title
    plt.xlabel('Steps')
    plt.ylabel('Rouge Score')
    plt.title('Rouge Scores')

    # Display legend
    plt.legend()

    # Show the plot
    plt.show()
```
